```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

### Problem Statement:

- The Indian bike market offers numerous options with different prices, mileage, engine capacities, and ratings, making it difficult for customers to choose the best bike. This project aims to collect bike data from BikeWale through web scraping and analyze key features. The goal is to identify and rank the best-performing bikes using data-driven insights.

```python
url = "https://www.bikewale.com/new-bike-search/?page=1"
url
```

```python

bike_name = []
rating = []
total_ratings = []
engine_cc = []
mileage = []
power = []
weight = []
price = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

for i in range(1,15):

    url = f"https://www.bikewale.com/new-bike-search/?page={i}"

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    cards = soup.find_all("h3", class_="o-c6")

    print(f"Page {i}: {len(cards)}")

    for card in cards:

        text = card.parent.parent.parent.get_text(" | ", strip=True)

        parts = [x.strip() for x in text.split("|") if x.strip()]

        bike_name.append(parts[0])

        rating.append(parts[1] if len(parts) > 1 else None)

        total_ratings.append(
            next((x for x in parts if "Ratings" in x), None)
        )

        engine_cc.append(
            next((x for x in parts if "cc" in x), None)
        )

        mileage.append(
            next((x for x in parts if "kmpl" in x), None)
        )

        power.append(
            next((x for x in parts if "bhp" in x), None)
        )

        weight.append(
            next((x for x in parts if "kg" in x), None)
        )

        price.append(
            next((x for x in parts if "₹" in x), None)
        )

```

```python
print(response)
print(response.status_code)
print(response.url)
print(response.text[:500])
```

```python
print(len(bike_name))
print(len(rating))
print(len(total_ratings))
print(len(engine_cc))
print(len(mileage))
print(len(power))
print(len(weight))
print(len(price))
```

```python
df = pd.DataFrame({
    "Bike_Name": bike_name,
    "Rating": rating,
    "Total_Ratings": total_ratings,
    "Engine_CC": engine_cc,
    "Mileage": mileage,
    "Power_BHP": power,
    "Weight_KG": weight,
    "Price": price
})

print(df.shape)
df.head()
```
